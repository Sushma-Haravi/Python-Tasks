```python
def sep_datatypes(data):
    cat=list(data.select_dtypes(include='object').columns)
    num=list(data.select_dtypes(exclude='object').columns)
    print(tabulate({"categorical":cat,"numerical":num},headers='keys'))
sep_datatypes(data)
```

```python
def info_of_cat(col):
    print(f" the unique values in {col} is : {data[col].unique()}")
    print(f" the mode of the {col} is : {data[col].mode()[0]}")
    print(f" the missing values in {col} is : {data[col].isnull().sum()}")
```

```python
def info_of_num(col):
    print(f" mean of the {col} is : {data[col].mean()}")
    print(f" mode of the {col} is : {data[col].mode()[0]}")
    print(f" median of the {col} is : {data[col].median()}")
    print(f" standard deviation of {col} is : {data[col].std()}")
    print(f" missing values in {col} is : {data[col].isnull().sum()}")
```

```python
def bar_percentage(ax,count='number of rows in data'):
    for bar in ax.patches:
        percentage=f"{round((bar.get_height()/count)*100,2)}%"
        
        x=bar.get_x()+bar.get_width()/2
        y=bar.get_height()
        ax.annotate(percentage,(x,y),va="bottom",ha="center")
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,6))
plt.title('count of defaulters and non defaulters')
count=data['TARGET'].value_counts()
print(count)
label=['Others','Defaulters']
ax=sns.countplot(x=data.TARGET,ax=ax[0])
bar_percentage(ax,len(data))
plt.pie(count,labels=label,autopct='%0.2f%%')

plt.show()
```
